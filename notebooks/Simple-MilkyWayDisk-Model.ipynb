# Expected observed kinematics for a simple Milky Way disk model

Calculate the expected sky radial velocities and proper motions for stars in an extremely simple model of the kinematics of the Milky Way disk. This is intended to roughly reproduce the motions of stars measured with respect to the solar system barycentre as a function of position on the sky and distance. The application is to fit models of the Milky Way rotation curve to Gaia astrometry and radial velocities.

## Milky Way toy kinematic model

In this model the Milky Way is in strict cylindrical differential rotation, with all stars on circular orbits around the Galactic centre, moving at speeds $V_{\mathrm{c},\star}=V_\mathrm{c}(R_\star)$, where $V_\mathrm{c}(R)$ is the circular speed which depends only the cylindrical radial coordinate $R$. The Sun is assumed to move at speed $V_{\mathrm{c},\odot} = V_\mathrm{c}(R_\odot)$, the circular speed at the distance from the sun to the Galactic centre. This model can obviously be approximately correct only in the Galactic plane. The following schematic diagram shows the steps to calculate the radial velocity $v_\mathrm{rad}$ of the star with respect to the sun, __assuming both the sun and the star are exactly in the plane of the disk__.

The circular speed is a positive quantity $V_\mathrm{c}(R) = R|\dot\phi|$, where $\dot\phi$ is positive for counterclockwise rotation in the righthanded coordinate system in the diagram below.

![Toy model](DiskKinematicsVrad.png)

The sun and the star are located at $\mathbf{x}_\odot$ and $\mathbf{x}$ and moving with velocities $\mathbf{v}_\odot$ and $\mathbf{v}$. The Sun is located at $(-R_\odot,0,0)$. Expressed in terms of the cylindrical coordinates we have:

\begin{equation}
    \mathbf{x}_\odot = \begin{pmatrix} -R_\odot \\ 0 \\ 0 \end{pmatrix}\,, \quad
    \mathbf{x} = \begin{pmatrix} R\cos\phi \\ R\sin\phi \\ 0 \end{pmatrix}\,, \quad
    \mathbf{v}_\odot = \begin{pmatrix} 0 \\ V_\mathrm{c}(R_\odot) \\0 \end{pmatrix}\,, \quad
    \mathbf{v} = \begin{pmatrix} V_\mathrm{c}(R)\sin\phi \\ -V_\mathrm{c}(R)\cos\phi \\0 \end{pmatrix}
\end{equation}

Thus note that the Milky Rotates clockwise in the diagram above, which is a _negative_ sense of rotation. The radial velocity of the star with respect to the sun is given by:

\begin{equation}
    v_\mathrm{rad} = \frac{\mathbf{x}-\mathbf{x}_\odot}{d} \cdot (\mathbf{v}-\mathbf{v}_\odot)\,,
\end{equation}

which works out to

\begin{equation}
    v_\mathrm{rad} = \frac{1}{d} \begin{pmatrix} R\cos\phi+R_\odot \\ R\sin\phi \\ 0 \end{pmatrix} \cdot
    \begin{pmatrix} V_\mathrm{c}(R)\sin\phi \\ -V_\mathrm{c}(R)\cos\phi-V_\mathrm{c}(R_\odot) \\ 0 \end{pmatrix} =
    \frac{1}{d}\left(V_\mathrm{c}(R)R_\odot - V_\mathrm{c}(R_\odot)R \right)\sin\phi =
    \frac{R}{d}\left(V_\mathrm{c}(R)\frac{R_\odot}{R} - V_\mathrm{c}(R_\odot)\right)\sin\phi\,.
\end{equation}

From the drawing above it can be seen that $\sin\phi = (d\sin\ell)/R$ and $\cos\phi=(d\cos\ell-R_\odot)/R$ so that in the end we obtain:

\begin{equation}
    v_\mathrm{rad} = \left(V_\mathrm{c}(R)\frac{R_\odot}{R} - V_\mathrm{c}(R_\odot)\right)\sin\ell\,,
\end{equation}

where

\begin{equation}
    R = \left(d^2+R_\odot^2-2dR_\odot\cos\ell\right)^{1/2}\,.
\end{equation}

For the proper motions in $\ell$ and $b$ a similar calculation leads to:

\begin{equation}
    \mu_{\ell*} = \frac{\cos\ell}{A_\mathrm{v}d}
    \left(V_\mathrm{c}(R)\frac{R_\odot}{R} - V_\mathrm{c}(R_\odot)\right) - \frac{V_\mathrm{c}(R)}{A_\mathrm{v}R}
    \,,\quad
     \mu_b = 0\,,
\end{equation}

where $A_\mathrm{v}=4.74047...$ km yr s$^{-1}$ and distance is in kiloparsec and velocity in km s$^{-1}$, resulting in proper motions in mas yr$^{-1}$.

The above equations hold whatever the values of $\ell$ and $d$. Note the expression for $\mu_b$ is zero for stars in the mid-plane of the disk.

### Expression for stars off the mid-plane

For stars located off the mid-plane (at some height $z$ or, equivalently, latitude $b$) the expressions should be calculated taken into account that for calculating $R$ the distance in the plane is needed. So if $d$ is still the distance to the star, the values of $R$, $\sin\phi$, and $\cos\phi$ are to be calculated using $d\cos b$ instead $d$ in the expression for $R$ above. The resulting expression for the radial velocity and proper motions are

\begin{equation}
    \begin{aligned}
        v_\mathrm{rad} & = \left(V_\mathrm{c}(R)\frac{R_\odot}{R} - V_\mathrm{c}(R_\odot)\right)\sin\ell\cos b \\
        \mu_{\ell,*} & = \frac{\cos\ell}{A_\mathrm{v}d}
        \left(V_\mathrm{c}(R)\frac{R_\odot}{R} - V_\mathrm{c}(R_\odot)\right) - \frac{V_\mathrm{c}(R)}{A_\mathrm{v}R}\cos b \\
        \mu_b & = \frac{1}{A_\mathrm{v}d}
        \left(V_\mathrm{c}(R_\odot) - V_\mathrm{c}(R)\frac{R_\odot}{R}\right)\sin\ell\sin b
    \end{aligned}\,.
\end{equation}

## Approximations for $d\ll R_\odot$ and the Oort parameters

Now observations of stars near the sun are considered, where "near" means that $d/R_\odot\ll 1$. From the expression for $R$ above one can write

\begin{equation}
    \frac{R_\odot}{R} = \left(1+\left(\frac{d}{R_\odot}\right)^2-2\frac{d}{R_\odot}\cos\ell \right)^{-1/2} \approx 1+\frac{d}{R_\odot}\cos\ell\,,
\end{equation}

where higher order terms in $d/R_\odot$ are ignored and the Taylor expansion for $(1-x)^{-1/2}$ was used. The expression for $v_\mathrm{rad}$ then becomes

\begin{equation}
    v_\mathrm{rad} \approx \left(V_\mathrm{c}(R)\left(1+\frac{d}{R_\odot}\cos\ell\right) - V_\mathrm{c}(R_\odot)\right)\sin\ell\,.
\end{equation}

We now make a further approximation by expanding to first order the circular velocity curve around its value at $R_\odot$.

\begin{equation}
    V_\mathrm{c}(R) \approx V_\mathrm{c}(R_\odot) + (R-R_\odot)\frac{\partial V_\mathrm{c}}{\partial R}\,,
\end{equation}

where the partial derivative is evaluated at $R_\odot$. Substituting into the expression for $v_\mathrm{rad}$ leads to

\begin{equation}
    \begin{aligned}
        v_\mathrm{rad} \approx & \left(\left(V_\mathrm{c}(R_\odot)+R\left(1-\frac{R_\odot}{R}\right)\frac{\partial V_\mathrm{c}}{\partial R}\right) 
        \left(1+\frac{d}{R_\odot}\cos\ell\right) - V_\mathrm{c}(R_\odot)\right)\sin\ell \\
        \approx & \left(\left(V_\mathrm{c}(R_\odot)-R\frac{d}{R_\odot}\cos\ell\frac{\partial V_\mathrm{c}}{\partial R}\right) 
        \left(1+\frac{d}{R_\odot}\cos\ell\right) - V_\mathrm{c}(R_\odot)\right)\sin\ell \\
        \approx & \left(\frac{V_\mathrm{c}(R_\odot)}{R_\odot}d\cos\ell - \frac{R}{R_\odot}\frac{\partial V_\mathrm{c}}{\partial R}d\cos\ell\right)\sin\ell \\
        \approx & \left(\frac{V_\mathrm{c}(R_\odot)}{R_\odot}d\cos\ell - \frac{\partial V_\mathrm{c}}{\partial R}d\cos\ell\right)\sin\ell 
        = \frac{1}{2}\left(\frac{V_\mathrm{c}(R_\odot)}{R_\odot} - \frac{\partial V_\mathrm{c}}{\partial R}\right)d\sin(2\ell)\\
    \end{aligned}\,,
\end{equation}

where in the last step we used that $R/R_\odot \approx 1-(d/R_\odot)\cos\ell$ and ignored terms with $d^2$. If we now define the Oort $A$ parameter as:

\begin{equation}
    A = \frac{1}{2}\left(\frac{V_\mathrm{c}(R)}{R} - \frac{\partial V_\mathrm{c}}{\partial R}\right)_{R=R_\odot}\,,
\end{equation}

we can write

\begin{equation}
    v_\mathrm{rad} \approx Ad\sin(2\ell) \,.
\end{equation}

Thus for stars near the sun one expects a radial velocity pattern which varies with $\sin(2\ell)$ across the sky and is proportional to the distance to the stars in the sample.

### Effect of the sun's peculiar motion

In practice the sun might not follow exactly the velocit pattern dictated by $V_\mathrm{c}(R)$ but will have a small motion, called "peculiar motion" on top of the velocity resulting from the rotation of the disk. We can write

\begin{equation}
    \mathbf{v}_\odot = \begin{pmatrix} 0 \\ V_\mathrm{c}(R_\odot) \\0 \end{pmatrix} + \begin{pmatrix} U_\odot \\ V_\odot \\ W_\odot \end{pmatrix}\,,
\end{equation}

where $(U_\odot,V_\odot,W_\odot)$ are the components of the sun's motion along $x$, $y$, and $z$. Substituting this into the first expression for $v_\mathrm{rad}$ above results in the addition of three components to the radial velocity that only depend on the celestial position of the star. The complete expression for the radial velocity then becomes

\begin{equation}
    v_\mathrm{rad} \approx Ad\sin(2\ell) -U_\odot\cos\ell -V_\odot\sin\ell \,,
\end{equation}

which holds for stars in the mid-plane of the disk, with sun also located in the mid-plane.

### Complete expressions for the observed proper motions when $d\ll R_\odot$

Applying the same approximation to the proper motions leads to the introduction of a second Oort parameter $B$

\begin{equation}
    B = \frac{1}{2}\left(-\frac{V_\mathrm{c}(R)}{R} - \frac{\partial V_\mathrm{c}}{\partial R}\right)_{R=R_\odot}\,.
\end{equation}

The complete expressions for the proper motions are

\begin{equation}
    \begin{aligned}
        \mu_{\ell*} & \approx \frac{1}{A_\mathrm{v}d}\left(\left(B+A\cos(2\ell)\right)d + U_\odot\sin\ell - V_\odot\cos\ell\right) \\
        \mu_b & \approx -\frac{W_\odot}{A_\mathrm{v}d}
    \end{aligned}\,.
\end{equation}

If we consider stars outside the mid-plane of the Milky Way and assumed the still follow the same orbits as stars in the plane (which is reasonable for stars close to the mid-plane of the disk) we can write the expressions for the observed proper motions and radial velocities as:

\begin{equation}
    \begin{aligned}
        v_\mathrm{rad} & \approx Ad\sin(2\ell)\cos^2 b -U_\odot\cos b\cos\ell -V_\odot\cos b\sin\ell -W_\odot\sin b \\[5pt]
        \mu_{\ell*} & \approx  \frac{1}{A_\mathrm{v}d}\left(\left(B+A\cos(2\ell)\right)d\cos b + U_\odot\sin\ell - V_\odot\cos\ell\right) \\[5pt]
        \mu_b & \approx \frac{1}{A_\mathrm{v}d}\left(-A\sin(2\ell)d\cos b\sin b + (U_\odot\cos\ell + 
        V_\odot\sin\ell)\sin b - W_\odot\cos b \right)
    \end{aligned}\,.
\end{equation}

These expressions are obtained by using $d\cos b$ instead of $d$ as the in-plane distance in the expressions above, and where for the radial velocity an extra projection factor $\cos b$ should be included to account of the projection of the velocity on the line of sight. The factor $1/(A_\mathrm{v}d)$ for the proper motions remains the same as it is the full distance that determines the magnitude of the proper motion.

Keep in mind that in these expressions the terms involving the Oort parameters are approximate while the terms for the sun's peculiar motion are exact (they simply represent the sun's motion projected along the line of sight or along the directions of increasing $\ell$ and $b$).

## General method to calculate the observed radial velocities and proper motions

A more general method to obtain a complete description of the velocity field in a differentially rotating stellar disk is presented in [Brunetti & Pfenniger (2010)](https://ui.adsabs.harvard.edu/#abs/2010A&A...510A..34B/abstract). Here we use vector astrometric methods to achieve the same for the __observed__ velocity field projected on the sky. This method can be applied to any model of the Milky Way and is not restricted to the mid-plane of the disk, or to the sun being located in the mid-plane.

We now consider a star at distance $d$ and Galactic celestial coordinates $(\ell,b)$. To calculate the position, radial velocity, and proper motions of the star the normal triad $[\mathbf{p}, \mathbf{q}, \mathbf{r}]$ is used:

\begin{equation}
    \mathbf{p} = \begin{pmatrix} -\sin\ell \\ \cos\ell \\ 0 \end{pmatrix}\,,\quad
    \mathbf{q} = \begin{pmatrix} -\sin b\cos\ell \\ -\sin b\sin\ell \\ \cos b \end{pmatrix}\,,\quad
    \mathbf{r} = \begin{pmatrix} \cos b\cos\ell \\ \cos b\sin\ell \\ \sin b \end{pmatrix}\,.
\end{equation}

The star is located at:

\begin{equation}
    \mathbf{x} = d\mathbf{r}+\mathbf{x}_\odot
\end{equation}

Its radial velocity and proper motions are then given by:

\begin{equation}
    v_\mathrm{rad} = \mathbf{r}\cdot(\mathbf{v}-\mathbf{v}_\odot)\,,\quad
    \mu_{\ell*} = \frac{1}{A_\mathrm{v}d}\mathbf{p}\cdot(\mathbf{v}-\mathbf{v}_\odot)\,,\quad
    \mu_b = \frac{1}{A_\mathrm{v}d}\mathbf{q}\cdot(\mathbf{v}-\mathbf{v}_\odot)\,,
\end{equation}

## Select a Milky Way model and plot the rotation curve

Several rotation curves are implemented as extensions of the [gala](http://gala.adrian.pw/en/latest/index.html) `BovyMWPotential2014` class. The extensions can only be used to extract a rotation curve, they do not represent Milky Way potentials. The next code block can be used to choose from several rotation curves:
* A flat rotation curve for which $V_\mathrm{c}(R)$ is constant as a function of $R$
* A solid body rotation curve for which $V_\mathrm{c}(R) \propto R$
* The rotation curve of the MWPotential2014 from [galpy](https://galpy.readthedocs.io/en/latest/potential.html) and described in [Bovy (2015)](https://ui.adsabs.harvard.edu/#abs/2015ApJS..216...29B/abstract).
* The rotation curve from [Brunetti & Pfenniger (2010)](https://ui.adsabs.harvard.edu/#abs/2010A&A...510A..34B/abstract)

```python
Rsun = 8277.0*u.pc
#zsun = 20.8*u.pc
zsun = 16*u.pc
sunpos = np.array([-Rsun.value, 0, zsun.value])*u.pc
#vsunpeculiar = np.array([11.1, 12.24, 7.25])*u.km/u.s
vsunpeculiar = np.array([8.1, 11.2, 8.1])*u.km/u.s

#diskmodel = DiskKinematicModel(BovyMWPotential2014(), sunpos, vsunpeculiar)
#diskmodel = DiskKinematicModel(FlatRotationCurve(240.0), sunpos, vsunpeculiar)
#diskmodel = DiskKinematicModel(SolidBodyRotationCurve(240.0, Rsun), sunpos, vsunpeculiar)
#diskmodel = DiskKinematicModel(SlopedRotationCurve(234.0, Rsun.value, -3.6), sunpos, vsunpeculiar)
diskmodel = DiskKinematicModel(BrunettiPfennigerRotationCurve(234.0, Rsun.value, 3.0, -0.55), sunpos, vsunpeculiar)
oortA, oortB = diskmodel.get_oort_ab(np.array([-Rsun.value, 0, 0])*u.pc)
print(f"Oort parameters A and B: {oortA}, {oortB}")

rc('font', family='sans-serif', size=18)
fig, axrc = plt.subplots(1, 1, figsize=(8,6), tight_layout=True)

rstar = np.linspace(0.1, 16, 1000)*u.kpc
starpos = np.vstack([rstar, rstar*0, rstar*0])
vcirc = diskmodel.get_circular_velocity(starpos)
axrc.plot(rstar, vcirc)
axrc.set_xlabel(r'$R$ [kpc]')
axrc.set_ylabel(r'$V_\mathrm{circ}$ [km s$^{-1}$]')
plt.show()
```

## Compare approximate and exact expressions for the kinematics in the mid-plane of the disk

The following plots show (for zero solar peculiar motion) a comparison between the approximate expressions for the radial velocites and proper motions and the exact expressions derived aboce.

```python
distance = 2.1*u.kpc
galat = np.deg2rad(-5)
dist_plane = distance*np.cos(galat)
galon = np.linspace(0,2*np.pi,361)
Rstar = np.sqrt(dist_plane**2 + Rsun**2 -2*dist_plane*Rsun*np.cos(galon)).to(u.pc)
vcsun = diskmodel.get_circular_velocity(np.array([-Rsun.value, 0, 0])*u.pc)
qstar = np.vstack([-Rstar.value, np.zeros_like(galon), np.zeros_like(galon)+distance.to(u.pc).value*np.sin(galat)])*u.pc
vcstar = diskmodel.get_circular_velocity(qstar)

vradpec = -vsunpeculiar[0]*np.cos(galon)*np.cos(galat) -vsunpeculiar[1]*np.sin(galon)*np.cos(galat) -vsunpeculiar[2]*np.sin(galat)
vrad_approx = (distance*oortA*np.sin(2*galon)*np.cos(galat)**2).to(u.km/u.s) + vradpec
vrad_exact = (vcstar*Rsun/Rstar-vcsun)*np.sin(galon)*np.cos(galat) + vradpec

pmlpec =  (vsunpeculiar[0]*np.sin(galon) -vsunpeculiar[1]*np.cos(galon))/(au_km_year_per_sec*distance)
pml_approx = (1/au_km_year_per_sec)*(oortB+oortA*np.cos(2*galon))*np.cos(galat) + pmlpec
pml_exact = (np.cos(galon)/(au_km_year_per_sec*distance)*(vcstar*Rsun/Rstar-vcsun) - vcstar/(au_km_year_per_sec*Rstar)*np.cos(galat)) + pmlpec

pmbpec = (-vsunpeculiar[0]*np.cos(galon)*np.sin(galat) +vsunpeculiar[1]*np.sin(galon)*np.sin(galat) -vsunpeculiar[2]*np.cos(galat))/(au_km_year_per_sec*distance)
pmb_approx = (1/au_km_year_per_sec)*(-oortA*np.sin(2*galon)*np.sin(galat)*np.cos(galat)) + pmbpec
pmb_exact = 1/(au_km_year_per_sec*distance)*(vcsun-vcstar*Rsun/Rstar)*np.sin(galat)*np.sin(galon) + pmbpec

rc('font', family='sans-serif', size=18)

fig=plt.figure(constrained_layout=True, figsize=(14,15))
gs = GridSpec(3, 1, figure=fig)

axvrad = fig.add_subplot(gs[0,0])
axvrad.plot(np.rad2deg(galon), vrad_approx, label='Approximate')
axvrad.plot(np.rad2deg(galon), vrad_exact, label='Exact')
axvrad.set_xlabel('Galactic longitude [$^\circ$]')
axvrad.set_ylabel(r'$v_\mathrm{rad}$ [km s$^{-1}$]')
axvrad.legend(title=rf"$d={distance.value:.1f}$ kpc, $b={np.rad2deg(galat):.1f}^\circ$")

axpml = fig.add_subplot(gs[1,0])
axpml.plot(np.rad2deg(galon), pml_approx, label='Approximate')
axpml.plot(np.rad2deg(galon), pml_exact, label='Exact')
axpml.set_xlabel('Galactic longitude [$^\circ$]')
axpml.set_ylabel(r'$\mu_{\ell*}$ [mas yr$^{-1}$]')
axpml.legend(title=rf"$d={distance.value:.1f}$ kpc, $b={np.rad2deg(galat):.1f}^\circ$")

axpmb = fig.add_subplot(gs[2,0])
axpmb.plot(np.rad2deg(galon), pmb_approx, label='Approximate')
axpmb.plot(np.rad2deg(galon), pmb_exact, label='Exact')
axpmb.set_xlabel('Galactic longitude [$^\circ$]')
axpmb.set_ylabel(r'$\mu_b$ [mas yr$^{-1}$]')
axpmb.legend(title=rf"$d={distance.value:.1f}$ kpc, $b={np.rad2deg(galat):.1f}^\circ$")

plt.show()
```

## Calculate and plot the differential velocity field

The following plots show the velocity field in galactocentric Cartesian coordinates in the mid-plane of the disk.

```python
grid_extent = [-15,15,-15,15]
nx = int(10*(grid_extent[1]-grid_extent[0]))
ny = int(10*(grid_extent[3]-grid_extent[2]))
xdef = np.linspace(grid_extent[0], grid_extent[1], nx)
ydef = np.linspace(grid_extent[2], grid_extent[3], ny)
xx, yy = np.meshgrid(xdef, ydef, indexing='xy') << u.kpc
zz = 0.0*u.kpc
sl = (slice(0,-1,10), slice(0,-1,10))

pml, pmb, vrad, vtan, p, q, r = diskmodel.differential_velocity_field(xx, yy, zz)

xsl = xx[sl]
ysl = yy[sl]
vradx = (r[0,:,:]*vrad)[sl]
vrady = (r[1,:,:]*vrad)[sl]
pmlx = (p[0,:,:]*pml)[sl]
pmly = (p[1,:,:]*pml)[sl]
```
